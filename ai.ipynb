```python
import pandas as pd

season = "2021-2022"
str = "https://fbref.com/en/comps/9/{}/stats/{}-Premier-League-Stats"

# Combining the URL and season strings

url = str.format(season, season)
url
```

The above code modularises the URL, so that various different seasons can be used, as seen below.

```python
seasons = ["2018-2019", "2019-2020", "2020-2021", "2021-2022"]
str = "https://fbref.com/en/comps/9/{}/stats/{}-Premier-League-Stats"

for season in seasons:
    url = str.format(season, season)
    print(url)
```

```python
df = pd.read_html(url, header = 0)
df[1]
```
